```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 100)
```

```python
data = pd.read_csv('average_rental_price_per_gemeente.csv', delimiter = ',')
average_rental_price_per_gemeente = pd.DataFrame(data)

data = pd.read_csv('SESWOA_and_avg_Housing_price.csv', delimiter = ',')
SESWOA_and_avg_Housing_price = pd.DataFrame(data)
```

```python
average_rental_price_per_gemeente.head()
```

```python
SESWOA_and_avg_Housing_price.head(5)
```

```python
SESWOA_and_avg_Housing_price_2019 = SESWOA_and_avg_Housing_price[SESWOA_and_avg_Housing_price['Periode'].str.contains('2019')].copy()
SESWOA_and_avg_Housing_price_2019['Periode'] = SESWOA_and_avg_Housing_price_2019['Periode'].str[:4]
```

```python
SESWOA_and_avg_Housing_price_2019.head(5)
```

```python
average_rental_price_per_gemeente.head()
```

```python
complete_table = pd.merge(
    SESWOA_and_avg_Housing_price_2019,
    average_rental_price_per_gemeente,
    left_on='Gemeente',
    right_on='Gemeente',
    how='left'
)
```

```python
complete_table
```

```python
relevant_columns = ['Periode', 'SES-Score', 'Spreiding', 'Gemeente', 'Region_Code', 'GemiddeldeVerkoopprijs_1', 'Average_Room_Price']

Final_table = complete_table[relevant_columns]
order = ['Periode', 'Gemeente', 'Region_Code', 'SES-Score', 'Spreiding', 'GemiddeldeVerkoopprijs_1', 'Average_Room_Price']
Final_table = Final_table[order]
```

```python
Final_table.to_csv('Final_table.csv')
```

```python
SESWOA_raw = final_merged_data[
    (final_merged_data['MeasureGroupId'] == 'M007817_2') &
    (final_merged_data['Title'] == 'Gemiddelde score')
]
SESWOA_raw
```

```python
SESWOA = SESWOA_raw[['Perioden','Value','Title_x','Title_y','DetailRegionCode','Description_x']].copy()

SESWOA['Wijk'] = np.nan
SESWOA['Buurt'] = np.nan
columns_order = ['Perioden','Value','Title_x','Buurt','Wijk','Title_y','DetailRegionCode','Description_x']
SESWOA = SESWOA[columns_order]
SESWOA.columns = [
    'Period',
    'Value',
    'Plaats',
    'Buurt',
    'Wijk',
    'Gemeente',
    'Region_Code',
    'Region_Description'
]
```

```python
import re

wijkcodes = {}

def update_wijken(row):
    if row['Region_Code'].startswith('WK'):
        row['Wijk'] = re.sub(r'^Wijk \d+\s+', '', row['Plaats'])
        wijkcodes[row['Region_Code'][2:8]] = row['Wijk']
    return row

def update_buurten(row):
    if row['Region_Code'].startswith('BU'):
        row['Buurt'] = row['Plaats']
        row['Wijk'] = re.sub(r'^Wijk \d+\s+', '', wijkcodes[row['Region_Code'][2:8]])
    return row

SESWOA = SESWOA.apply(update_wijken, axis=1)
SESWOA = SESWOA.apply(update_buurten, axis=1)
```

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 100)
```

```python
data = pd.read_csv('SES-WOA Score (CBS)/SESWOA_scores.csv', delimiter = ',')
SESWOA_scores = pd.DataFrame(data)

data = pd.read_csv('Housing Prices (CBS)/housing_price.csv', delimiter = ';')
housing_prices = pd.DataFrame(data)
```

```python
SESWOA_scores_gemeente_only = SESWOA_scores[SESWOA_scores['Region_Code'].str.startswith('GM')]
```

```python
SESWOA_and_avg_Housing_price = pd.merge(
    SESWOA_scores_gemeente_only,
    housing_prices,
    left_on=['Region_Code', 'Periode'],
    right_on=['RegioS', 'Perioden'],
    how='inner'
)
SESWOA_and_avg_Housing_price
```

```python
SESWOA_and_avg_Housing_price.to_csv('SESWOA_and_avg_Housing_price.csv', index = False)
```

```python
def preprocess_inputs(df):
    df = df.copy()
    
    # Drop bad rows
    bad_rows = df.query("crawlStatus == 'unavailable'").index
    df = df.drop(bad_rows, axis=0).reset_index(drop=True)
    
    # Use only select features
    df = df[[
        'areaSqm',
        'city',
        'furnish',
        'latitude',
        'longitude',
        'propertyType',
        'rent',
        'internet',
        'kitchen',
        'living',
        'pets',
        'shower',
        'smokingInside',
        'toilet'        
    ]]
    return df
```

```python
Y = preprocess_inputs(data)
```

```python
Y.isna().sum()
```

```python
Y2 = Y.replace({ '': np.NaN, 'Unknown': np.NaN}) 
```

```python
{column: list(Y2[column].unique()) for column in Y.select_dtypes('object').columns}
```

```python
Y2.isna().mean()
```

```python
missing_values = Y2.columns[Y2.isna().sum() > 0]
for column in missing_values: 
    Y2[column] = Y2[column].fillna(Y2[column].mode()[0])
```

```python
def preprocess_for_dates(df):
    # Create an explicit copy of the relevant columns
    df = df[[
        '_id',
        'externalId',
        'city',
        'datesPublished',
        'firstSeenAt',
        'lastSeenAt'
    ]].copy()  # Explicit copy to avoid SettingWithCopyWarning

    # Convert firstSeenAt and lastSeenAt to datetime
    df['firstSeenAt'] = pd.to_datetime(df['firstSeenAt'].apply(lambda x: x['$date']))
    df['lastSeenAt'] = pd.to_datetime(df['lastSeenAt'].apply(lambda x: x['$date']))
    return df
```

```python
Y_dates = preprocess_for_dates(data)
Y_dates_sorted = Y_dates.sort_values(by='lastSeenAt', ascending=False)
Y_dates_sorted
```
